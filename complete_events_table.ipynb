In this notebook, we update `Event Definitions.csv` so that it contains all event ids

```python
import numpy as np
import pandas as pd
```

```python
events = pd.read_csv("data/Event Definitions.csv")
```

```python
events.head()
```

```python
actions = pd.read_csv("data/cleaned_dat_train1.csv")
```

```python
actions.head()
```

```python
new_event_example = actions[actions['ed_id'] == 1].iloc[0]
```

```python
new_event_example
```

```python
# There are event ids in the range from 1 to 29
current_ids = events['event_definition_id'].unique()

for event_id in range(1, 30):
    if event_id not in current_ids:
        print(f"Adding row for event id {event_id}")
        new_event_example = actions[actions['ed_id'] == event_id].iloc[0]
        new_event = pd.DataFrame([{
            "event_name": new_event_example['event_name'],
            "journey_id": 1,
            "event_definition_id": event_id,
            "milestone_number": np.nan,
            "stage": np.nan
        }])
        events = pd.concat([events, new_event], ignore_index=True)
```

```python
events = events.sort_values(by='event_definition_id').reset_index(drop=True)
```

```python
events.head()
```

```python
events.to_csv("data/complete_event_definitions.csv", index=False)
```
